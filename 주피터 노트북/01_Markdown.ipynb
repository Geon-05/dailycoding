# Markdown 사용하기

### 😎 Header -> '#'은 6개까지만!
# Header 1 : # 1개 사용
## Header 2 : # 2개 사용
### Header 3 : # 3개 사용
#### Header 4 : # 4개 사용
##### Header 5 : # 5개 사용
###### Header 6 : # 6개 사용
- HTML 문법 h1~6으로 사용가능!

### 😎 Quote(인용구) -> '>'
> 인용구는 '>'로 사용 가능!
>> 인용구 안에 인용구 를 넣을 수 있다.

### 😎 강조(굵은글씨) -> '\*\*문자열\*\*'
**강조문자** 일반문자

### 😎 강조(기울임) -> '\*문자열\*'
*강조문자* 일반문자

### 😎 수학기호 사용 => '\$수학기호\$'
$y=3x$

### 😎 List(목록) -> 순서O '1./2./3.' 순서X '*/+/-'
### 들여쓰기는 Tab으로 구분!
1. 첫 번째
2. 두 번째
3. 세 번째
* 목록
+ 목록
- 목록
    1. 들여 첫 번째
    2. 들여 두 번째
    * 들여 목록

### 😎 수평선 -> '***' / '---'
***
---

### 😎 줄바꿈 -> space를 두 번 후 enter처리 or HTML문법 br
줄바꿔!  
바꿨어!<br>
HTML문법도 가능!
